```python
import json
import spacy
```

```python
from spacy import displacy
```

```python
# load the training data
with open('/content/drive/MyDrive/training_data.json') as fp:
  training_data = json.load(fp)
```

```python
training_data
```

```python
# prepare an empty model to train
nlp = spacy.blank('en')
nlp.vocab.vectors.name = 'demo'
ner = nlp.create_pipe('ner')
nlp.add_pipe(ner, last = True)
```

```python
# Add the custome NER Tags as entities into the model
for label in training_data["classes"]:
  nlp.entity.add_label(label)
```

```python
optimizer = nlp.begin_training()
```

```python
for iter in range(30):
  losses = {}
  for text, annotations in training_data["annotations"]:
    if len(text) > 0:
      nlp.update([text], [annotations], sgd=optimizer, losses = losses)
      print(losses)
```

```python
from pathlib import Path
output_dir=Path('/content/drive/MyDrive/ner spacy')
```

```python
nlp.meta['name'] = 'my_ner'  # rename model
nlp.to_disk(output_dir)
print("Saved model to", output_dir)
```

```python
nlp = spacy.load(output_dir)
```

```python
my_text = 'Bachelor of Technology in Information Technology from Bengal College of Engineering and Technology affiliated to University and  passed out in the year 2016 with 75 percentage.'
```

```python
doc = nlp(my_text)
```

```python
displacy.render(doc,jupyter=True, style = "ent")
```

```python
n=len(my_text)
```

```python
my_list = []
i = 0
while i < n :
  string = my_text[i]
  doc=nlp(string)
  print(string)
  Label = []
  Text = []
  for X in doc.ents:
    Label.append(X.label_)
    Text.append(X.text)
    Dict = {Label[n]: Text[n] for n in range(len(Label))}
  my_list.append(Dict)
  i = i+1
```
